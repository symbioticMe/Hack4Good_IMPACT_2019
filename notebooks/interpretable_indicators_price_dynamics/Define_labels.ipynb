```r
#months <- unique(aggregated_data$Month)[3 + 3*1:6]
months <- unique(aggregated_data$Month)[3:length(unique(aggregated_data$Month))]

new_label_df = list()
for (month in months){
    df = get_aggregated_table(aggregated_data,final_month = as.character(month), time_window=6,
                         geo_level = 'q_district', commodity_type = 'smeb_total_float')
    

    label_df = get_label(df, geo_level = 'q_district', commodity_type = 'smeb_total_float')
    
    label_df$Month=month
    new_label_df[[month]] = label_df
}

label_df_complete = do.call(rbind, new_label_df)

```

```r
head(label_df_complete)
```

```r
label_df_complete = label_df_complete %>%
   mutate(rel_mean = mean_der/mean_price,
          CV_der = sd_der/mean_price) %>%
filter(CV_der < 4)
ggplot(label_df_complete, aes(x=Month, y=CV_der))+geom_point()+geom_line()+facet_wrap(~q_district)+scale_y_log10()

#from this plot we can see that everything with more than 1 is volatile
```

```r
ggplot(label_df_complete, aes(x = CV_der))+geom_histogram()+scale_x_log10()
```

```r
ggplot(label_df_complete, aes(x = rel_mean))+geom_histogram()
```

```r
ggplot(label_df_complete %>% filter(rel_mean > - .05), aes(x=Month, y=rel_mean))+geom_point()+geom_line()+facet_wrap(~q_district)+
geom_hline(yintercept = 0, linetype = 'dashed')+theme_bw()
```

```r
ggplot(label_df_complete %>% filter(rel_mean > - .05) %>% filter(CV_der < .5), 
       aes(x=rel_mean, y = CV_der))+geom_point()+scale_y_log10()+
geom_vline(xintercept = 0, linetype = 'dashed')+theme_bw()+xlim(c(-.075, 0.075))+
geom_hline(yintercept = .075, color = 'forestgreen', linetype = 'dashed')+
geom_vline(xintercept = 0.01, linetype = 'dashed', color = 'blue')+
geom_vline(xintercept = -0.01, linetype = 'dashed', color = 'blue')
```

```r
ggplot(label_df_complete %>% filter(rel_mean > - .05) %>% filter(CV_der < .5), 
       aes(x=abs(rel_mean), y = CV_der, color = Month))+geom_point()+scale_y_log10()+
geom_vline(xintercept = 0, linetype = 'dashed')+theme_bw()+xlim(c(0, 0.075))+
geom_hline(yintercept = .05, color = 'forestgreen', linetype = 'dashed')+
geom_vline(xintercept = 0.01, linetype = 'dashed', color = 'blue')+
geom_vline(xintercept = -0.01, linetype = 'dashed', color = 'blue')
```
