```r
aggregated_data <- read.csv(file="data/aggregated_monthly_1.csv", header=TRUE, sep=",")
```

```r
source('src/get_aggregated_table.R')
df_aggregated = get_aggregated_table(aggregated_data,
                                     final_month = "Apr 19", 
                                     time_window=length(unique(aggregated_data$Month)),
                         geo_level = 'q_district', commodity_type = 'smeb_sanswater_float')
```

```r
dim(df_aggregated)
names(df_aggregated)
head(df_aggregated)
```

```r
geo_table = df_aggregated %>%
select(region, q_gov, q_district) %>% distinct() %>% 
filter(region != '') %>%
filter(!(q_district == 'SY0301' & region == 'besieged'))
```

```r
#months <- unique(aggregated_data$Month)[3 + 3*1:6]
months <- unique(aggregated_data$Month)[3:length(unique(aggregated_data$Month))]

new_label_df = list()
for (month in months){
    source("src/get_aggregated_table.R")
    df = get_aggregated_table(aggregated_data,final_month = as.character(month), time_window=6,
                         geo_level = 'q_district', commodity_type = 'smeb_sanswater_float')
    
    source("src/get_label_function.R")
    label_df = get_label(df, geo_level = 'q_district', commodity_type = 'smeb_sanswater_float')
    
    label_df$Month=month
    new_label_df[[month]] = label_df
}

label_df_complete = do.call(rbind, new_label_df)


```

```r
label_df_complete_missing_summary = label_df_complete %>%
group_by(Month) %>%
summarize(n_missing_SD = sum(is.na(sd_der)),
         n_missing_mean = sum(is.na(mean_der)))
label_df_complete_missing_summary
```
