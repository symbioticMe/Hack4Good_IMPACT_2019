```r
df_aggregated = get_aggregated_table(aggregated_data,
                                     final_month = "Apr 19", 
                                     time_window=length(unique(aggregated_data$Month)),
                         geo_level = 'q_district', commodity_type = 'smeb_total_float')
```

```r
dim(df_aggregated)
names(df_aggregated)
head(df_aggregated)
```

```r
#months <- unique(aggregated_data$Month)[3 + 3*1:6]
months <- unique(aggregated_data$Month)[3:length(unique(aggregated_data$Month))]

new_label_df = list()
for (month in months){
    df = get_aggregated_table(aggregated_data,final_month = as.character(month), time_window=6,
                         geo_level = 'q_district', commodity_type = 'smeb_total_float')
    
    label_df = get_label(df, geo_level = 'q_district', commodity_type = 'smeb_total_float')
    
    label_df$Month=month
    new_label_df[[month]] = label_df
}

label_df_complete = do.call(rbind, new_label_df)


```
