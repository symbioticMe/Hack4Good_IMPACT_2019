```python
import pandas as pd
from datetime import datetime as dt
import numpy as np
```

## Preprocessing

```python
smeb_df= pd.read_csv('SMEB_sbd_level 2019-10-29 .csv')
smeb_df.head()
```

```python
smeb_df.columns
```

Removing price columns and changing date format so that it follows the calender order and not treat it as a string

```python
smeb_only_df = smeb_df.copy()

for col in smeb_df.columns:
    if ("price" in col) or ("nonstandard" in col):
        smeb_only_df.drop(columns=[col], inplace=True)

smeb_only_df['Month'] = smeb_only_df['Month'].apply(lambda x : pd.to_datetime(x, format = "%B %y", errors='ignore'))
smeb_only_df['Month'] = smeb_only_df['Month'].apply(lambda x : pd.to_datetime(x, format = "%b %y", errors='ignore'))
```

We instead have a more readable format of high correlation counts

```python
def correlations(df, group, threshold = 0.8, col_start = 6):
    item_list = list(df.columns[col_start:])
    n = len(item_list)
    col_names = item_list
    df_corr = pd.DataFrame(columns=["Item 1", "Item 2", "Corr"])
    df_corr["Item 1"] = np.repeat(item_list, n)
    df_corr["Item 2"] = item_list*n
    for i in range(n):
        for j in range(i,n):
            if not i==j:
                c = sum(df.groupby(group)[item_list[i], item_list[j]].corr().iloc[0::2,-1]>=threshold)
                df_corr.iloc[i*n + j, 2] = c
    return df_corr
    
```

```python
df_corr = correlations(smeb_only_df, 'Month')
df_corr.sort_values("Corr", ascending=False).head(50)
```
